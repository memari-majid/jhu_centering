```python
import pandas as pd
import sqlalchemy
from matplotlib import pyplot as plt
import datetime
import copy
import numpy as np
```

# EMAs

```python
sleep = [1,4,7,10,13,16,19,22,25,28]
exercise = [2,5,8,11,14,17,20,23,26]
social = [3,6,9,12,15,18,21,24,27]
```

```python
df_sleep = pd.DataFrame()
df_sleep['Study ID'] = []
for i in sleep:
    df_temp = pd.read_csv('/data/meditation/majid/surveys/{}.csv'.format(str(i)), skiprows=[1,2])
    df_temp = df_temp[df_temp['Finished']==1]
    df_temp = df_temp.sort_values(by = 'EndDate', ascending = False)
    df_temp = df_temp.drop_duplicates(subset = "Study ID")
    df_temp = df_temp[['Study ID',
                       'ESAT_13',#happy
                       'ESAT_8',#sad
                        'Sleep_D{}'.format(str(i))
                         ]]
    df_temp['Study ID'] = df_temp['Study ID'].apply(lambda x: int(x))
    df_temp.columns = ['Study ID','Happy_D{}'.format(str(i)),'Sad_D{}'.format(str(i)),'Sleep_D{}'.format(str(i))]
    uniq_study_id = list(set(df_sleep['Study ID'].tolist()+df_temp['Study ID'].tolist()))
    df_merge_temp = pd.DataFrame()
    df_merge_temp['Study ID'] = uniq_study_id
    df_merge_temp = df_merge_temp.merge(df_sleep, on = 'Study ID', how = 'left')
    df_merge_temp = df_merge_temp.merge(df_temp, on = 'Study ID', how = 'left')
    df_sleep = copy.deepcopy(df_merge_temp)
    
df_sleep
    
    
```

```python
df_sleep.to_csv("survey_sleep.csv", index = False)
```
